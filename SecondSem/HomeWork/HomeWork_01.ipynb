```python
#-----------------------------
# Задание 1
#-----------------------------
class Node(object):
    def __init__(self, value=None, next_node=None, prev_node = None):
        self.value = value
        self.next = next_node
        self.prev = prev_node

    def __str__(self):
        return f'[Node with value {self.value}]'
    
def print_linked_list(head):
    cur = head
    while cur is not None:
        print(cur)
        cur = cur.next
    
def reverse_linked_list(head, tail=None):
    while head:
        head.next, tail, head = tail, head, head.next
    return tail

h, a, b, c, d = Node(1), Node(2), Node(3), Node('Внезапно'), Node(5)

h.next = a
a.next = b
b.next = c
c.next = d

print_linked_list(h)

h = reverse_linked_list(h)
print('-----')
print_linked_list(h)        
```

```python
#-----------------------------
# Задание 2
#-----------------------------
class Node(object):
    def __init__(self, value=None, next_node=None, prev_node = None):
        self.value = value
        self.next = next_node
        self.prev = prev_node
        
class LinkedList:
    def __init__(self):
        self.first = None
        
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.first is None:
            self.first = new_node
            return
        n = self.first
        while n.next is not None:
            n= n.next
        n.next = new_node; 
        
    def make_new_list(self, data):
        if len(data) == 0:
            return
        for i in range(0, len(data)):
            self.insert_at_end(data[i])
        
    def print_linked_list(self):
        cur = self.first
        while cur is not None:
            print(cur.value)
            cur = cur.next 
                  

    def bub_sort_linkchange(self):
        end = None
        while end != self.first:
            r = p = self.first
            while p.next != end:
                q = p.next
                if p.value > q.value:
                    p.next = q.next
                    q.next = p
                    if p != self.first:
                        r.next = q
                    else:
                        self.first = q
                    p,q = q,p
                r = p
                p = p.next
            end = p
            
h = LinkedList()
l = [2, 4, 1, 1, 2, 0]
h.make_new_list(l)

h.print_linked_list()
h.bub_sort_linkchange()
print('-----')
h.print_linked_list()
```
